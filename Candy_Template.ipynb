### 8 - clean ` country` column
> tip : note that country column has country names with different writing ways
  - use fuzzywuzzy to deal with this problem
  - you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
def rep(c) :
  if  str(c).isdecimal() :
    return np.nan
  return c
df["country"] = df["country"].apply(rep)

total = [ 'USA','United States of America', 'Canada', 'United Kingdom', 'UK', 'Australia', 'Germany', 'France', 'Italy', 'Spain']
def clean(c) :
  if c is np.nan :
    return c
  else :
    target, score = process.extractOne(c, total)
    if score > 60 :
      return target
    return c
df["country"] = df["country"].apply(clean)
```

```python
usa_vars = ['North Carolina ', 'Pittsburgh','New York','USA? Hard to tell anymore..','California','New Jersey','Alaska','USA','United States of America','Ahem....Amerca']
other = ['Europe','Earth','insanity lately','subscribe to dm4uz3 on youtube',"I don't know anymore", 'Fear and Loathing']

def usa(c) :
  if c in usa_vars :
    return "United States"
  elif c in other :
    return np.nan
  return c
df["country"] = df["country"].apply(usa)


df["country"].replace({'UK':'United Kingdom','endland':'United Kingdom', 'England' : 'United Kingdom'
, 'UD':'United Kingdom','hong kong' : "China", 'Hong Kong' : 'China', 'Scotland ':'Scotland','The Netherlands' : 'Netherlands', 'murrika': 'Spain'
}, inplace = True)

df['country'].fillna('unknown', inplace = True)
```

### 9 - clean ` area` column
> tip : area column has the same problem as country columns , but looking at all it's unique values may give us another way to deal with it rather than try to fix it

```python
# remove numeric entries
df['state'] = df['state'].apply(rep)
# a list by all states of america
USA_county = ["Alabama", "Alaska", "Arizona", "Arkansas",'Brooklyn', "California", "Colorado", "Connecticut", "Delaware", "Florida",
          "Georgia", "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana", "Maine",
          "Maryland", "Massachusetts", "Michigan", "Minnesota", "Mississippi", "Missouri", "Montana", "Nebraska", "Nevada",
          "New Hampshire", "New Jersey", "New Mexico", "New York", "North Carolina", "North Dakota", "Ohio", "Oklahoma", "Oregon","Pennsylvania",
  "Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Vermont", "Virginia", "Washington", "West Virginia", "Wisconsin", "Wyoming"]
other = ['USA','America','United States', 'Ontario']


def clean_county(c) : # match the variations of states of america in the data
  if c is np.nan :
    return c
  else :
    target, score = process.extractOne(c, USA_county+other)
    if score > 70 :
      return target
    return c
df['state'] = df['state'].apply(clean_county)

# replace the entries (USA, America, United States) to most frequent state of america in the data
most_frequent_state_in_USA = list(df.loc[df['state'].isin(USA_county), 'state'].value_counts())[0]
df['state'].replace('USA', most_frequent_state_in_USA)
df['state'].replace('America', most_frequent_state_in_USA)
df['state'].replace('United States', most_frequent_state_in_USA)

def twoChars_clean(c) : # a function to handle the 2 characters entries and replace them by one of america states
  if c is not np.nan and len(c) == 2 :
    a = c.lower()
    one = a[0]
    two = a[1]
    for state in USA_county :
      state_low = state.lower()
      if one in state_low and two in state_low and state_low.index(one) < state_low.index(two) :
        return state
  else :
    return c

df['state'] = df['state'].apply(twoChars_clean)
# filter entries having NaNs in state column and 'United States' in country column to replace their NaNs
df.loc[(df['state'].isna()) & (df['country'] == 'United States') , 'state'] = most_frequent_state_in_USA

```

```python
# find the countries with most nan values in state column
nan_counts = df['state'].isna().groupby(df['country']).sum().reset_index()
nan_counts.columns = ['country', 'nan_count']
nan_counts = nan_counts.sort_values(by='nan_count', ascending=False)
nan_counts
```

```python
# get the most frequent state in canada that revealed in the data and put it instead of nan for canada country
mostFreq_state_inCanada = list(df.loc[(df['country'] == 'Canada') & ~(df['state'].isna()), 'state' ].value_counts())[0]
df.loc[(df['country']== 'Canada') & (df['state'].isna())] = mostFreq_state_inCanada

# replace all remainig nans in state column by 'unknown'
df['state'].fillna('unknown', inplace = True)
```

### 10 - clean ` Q6` columns
>- tip : Q6 has all kinds of chocolate bars and questions with three distinct options to choose from (Joy, Meh, Despair).
- so all the column that has Q6 is a candy name , so you should go through all these column and you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
Q6_cols = [col for col in df.columns if 'Q6' in col]
# get rid of numbers
for col in Q6_cols :
  df[col] = df[col].apply(rep)

# drop the rows eith nan values more than half the length of Q6 columns
limit = len(Q6_cols) // 2
q6_df = pd.DataFrame(df[Q6_cols])
freq = list(q6_df.isna().sum(axis = 1))
for i ,f in enumerate(freq) :
  if f > limit :
    df.drop(i , inplace = True)
```

```python
 # Function to fill NaN values with the mode of the other Q6 columns in the same row
def clean_Q6(row):

  for col in Q6_cols:
      if pd.isna(row[col]) :
          non_nan_values = [row[q6_col] for q6_col in Q6_cols if q6_col != col and not pd.isna(row[q6_col])]

          if non_nan_values:
              mode_value = pd.Series(non_nan_values).mode()[0]
              row[col] = mode_value

  return row

df = df.apply(clean_Q6, axis=1)


```
